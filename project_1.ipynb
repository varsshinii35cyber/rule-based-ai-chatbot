```python
print("====================================")
print(" Welcome to DecodeLabs AI Chatbot ")
print("====================================")
print("Type 'exit' to end the chat.\n")

while True:
    user = input("You: ").lower()

    if user in ["hi", "hello", "hey"]:
        print("Bot: Hello! Welcome to DecodeLabs. How can I help you?")

    elif user == "how are you":
        print("Bot: I'm doing great! Thanks for asking.")

    elif user == "what is your name":
        print("Bot: I am a Rule-Based AI Chatbot.")

    elif user == "who made you":
        print("Bot: I was created as part of the DecodeLabs AI Internship.")

    elif user == "help":
        print("Bot: You can greet me, ask my name, ask how I am, or type 'exit' to quit.")

    elif user in ["bye", "exit", "quit"]:
        print("Bot: Goodbye! Have a great day.")
        break

    else:
        print("Bot: Sorry, I don't understand that. Please try another question.")
```
